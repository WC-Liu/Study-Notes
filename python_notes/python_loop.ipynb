打印99乘法表

```python
total = 0
for i in range(1, 10):
    for j in range(1, i+1):
        print(f'{j} x {i} = {i * j}' , end='\t')
    print()
```

输入直角边长，打印三角形

```python
l = int(input('输入直角边长: '))
for i in range(1, l+1):
    for j in range(1, i+1):
        print("*", end='\t')
    print()
```

打印国际象棋

```python
for i in range(8):
    for j in range(8):
        if (i+j) % 2 == 0:
            print("#", end=' ')
        else:
            print(" ", end=' ')
    print()
```
