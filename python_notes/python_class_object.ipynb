最小示例

```python
# 1. 定义一个类
class Phone:
    #构造方法： 创建对象时自动执行
    def __init__(self, brand, color):
        self.brand = brand  #属性： 品牌
        self.color = color  #属性： 颜色

    # 方法： 打电话
    def call(self):
        print(f"{self.color}的{self.brand}正在打电话")

# 2. 创建对象（实例化）
p1 = Phone("苹果", "白色")
p2 = Phone("华为", "黑色")

# 3. 使用对象
p1.call()
p2.call()
```
