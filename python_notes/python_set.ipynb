```python
lst = [1,2,2,3,3,3,4,4,4,4]
#集合去重
s = set(lst)
print(s)
#判断元素
print(1 in s)
print(5 in s)

#添加与删除
s1 = set()
s1.add(10)
s1.add(20)
s1.add(30)
print(s1)

s1.discard(20)
print(s1)
s1.clear()
print(s1)
```

集合运算

```python
a = {1,2,3,4}
b = {3,4,5,6}
print(a & b)
print(a | b)
print(a - b)
print(a ^ b)
```

```python
words = ['apple','banana','apple','orange','banana','apple']
s = len(set(words))
print(s)
```
