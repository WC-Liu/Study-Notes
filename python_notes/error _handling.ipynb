核心语法

```python
from numba.core.types import none

# 除零异常演示
try:
    result = 10 / 0  # 这行一定会报错
except ZeroDivisionError:
    print("错误：不能除以零！")
else:
    print("计算结果：", result)
finally:
    print("程序执行完毕")
```

用as变量名获取错误类型

```python
try:
    int("abc")
except ValueError as e: #约定俗成用as e
    print("捕获到异常：", e)  # 输出：invalid literal for int() with base 10: 'abc'
```

捕获多个异常

```python
# 写法1：分开捕获
try:
    num = int(input("输入数字："))
    print(10 / num)

except ValueError:
    print("请输入有效数字")
except ZeroDivisionError:
    print("不能输入0")

# 写法2：合并捕获
except (ValueError, ZeroDivisionError):
    print("输入错误")
```

异常捕获示例（文件读取）

```python
def read_file(filename):
    try:
        with open(filename, "r", encoding="utf-8") as f:
            content = f.read()
    except FileNotFoundError:
        print(f"错误： 文件{filename} 不存在")
        return None
    except Exception as e:
        print(f"未知错误： {e}")
        return None
    else:
        print("文件读取成功")
        return content
    finally:
        print("文件操作结束")

# 测试
read_file("test1.txt")
```
