写一个函数 add(a, b)，返回 a + b 的结果。

```python
def add(a, b):
    return a + b
print(add(3, 5))
```

写一个函数 square(x)，返回 x 的平方。

```python
def square(x):
    return x**2
print(square(3))
```

写一个函数 max2(a, b)，返回 a 和 b 中较大的数。

```python
def max2(a, b):
    return max(a, b)
print(max2(3, 5))
```

写一个函数 is_even(n)，判断 n 是否为偶数，是返回 True，否则 False。

```python
def is_even(num):
    return num % 2 == 0
print(is_even(1))
```

写一个函数 sum_n(n)，计算 1+2+…+n 的和并返回。

```python
def sum_n(n):
    total = 0
    for i in range(1, n + 1):
        total += i
    return total
print(sum_n(9))
```

写一个函数 count_words(s)，统计字符串 s 里有多少个单词（空格分隔）。

```python
def count_words(s):
    return len(s.split('，'))
print(count_words("我是，谁是我"))
```

写一个函数 reverse_str(s)，返回字符串 s 的反转。

```python
def reverse_str(s):
    return s[::-1]
print(reverse_str("abcdefg"))
```

写一个函数 avg(lst)，接收一个数字列表，返回平均值。

```python
def avg(lst):
    return sum(lst) / len(lst)
print(avg([1, 2, 3, 4, 5]))
```

写一个函数 find_max(lst)，返回列表中的最大值（不用内置 max）

```python
def find_max(lst):
    max1 = lst[0]  #不能定义为0，如果列表有负数
    for num in lst:
        if num > max1:
            max1 = num
    return max1
print(find_max([7, 5, 3, 9, 5]))
```
