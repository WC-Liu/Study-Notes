### 定义一个函数，用于根据传入的一批商品信息（商品名、价格、数量）、优惠（优惠券、积分抵扣）、运费信息计算订单的总金额。
#### 具体规则如下：
1. 优惠券需要商品金额满 5000 才可以使用，且优惠券金额不能超过商品总价。
2. 积分抵扣需要商品总金额满 5000 才可以使用，100 积分抵扣 1 元（且抵扣金额不能超过商品总价，积分只能整百抵扣）。

```python
def cal_order_cost(*args, yhq = 0, jf = 0, yf = 0):
    """

    :param args: 商品信息（名字、价格、数量）
    :param yhj: 优惠券
    :param jf: 积分
    :param yf: 运费
    :return: 订单总金额
    """
    #总金额 = 商品总金额 - 优惠 - 积分 + 运动费
    # 1. 计算商品总金额
    total_price = [goods[1] * goods[2] for goods in args]
    total_cost = sum(total_price)

    # 2. 扣减优惠券
    if total_cost >= 5000 and yhq < total_cost:
        total_cost -= yhq

    # 3. 扣减积分
    if total_cost >= 5000 and jf // 100 <= total_cost:
        total_cost -= jf // 100

    #4. 添加运费
    total_cost += yf

    return total_cost
total = cal_order_cost(("鼠标", 188, 3), ("键盘", 388, 1), ("手机", 6999, 1), yhq=1000, jf=5000, yf=12)
print(total)
```
