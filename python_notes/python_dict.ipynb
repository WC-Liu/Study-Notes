1.创建字典

```python
from abc import update_abstractmethods

#创建字典
person = dict(name = "张三", age = 18, class_ ="高三（1）班")
print(person)
#取出字典，[]和get两种方法
print(person["name"], person["age"])
print(person.get("name"))
print(person.get("height", "没有"))
#删除键值对
#del person["class_"]
#person.pop("class_")
print(person)
#遍历字典
for key in person:
    print(key)
for value in person.values():
    print(value)
for key, value in person.items():
    print(key, value)
```

6.统计字符次数
给定字符串s = "abacabc"，用字典统计每个字符出现的次数。

```python
s = "abacabc"
count = {}
for char in s:
    count[char] = count.get(char, 0) + 1
print(count)
```

7. 列表转字典
把两个列表keys = ["name", "age"]和values = ["李四", 20]转成字典。

```python
key = ["name", "age"]
values = ["李四", 20]
count2 = dict(zip(key, values))
print(count2)
```

8.字典推导式
生成一个字典：键是 1~10的偶数，值是键的 2 倍。

```python
dic = {x: x*2 for x in range(1, 11) if x % 2 == 0}
print(dic)
```

9. 按值排序
字典scores = {"张三":90, "李四":80, "王五":95}，按分数从高到低排序。

```python
scores = {"张三":90, "李四":80, "王麻子":95}
sorted_scores = dict(sorted(scores.items(), key=lambda x:x[1], reverse=True))
print(sorted_scores)
```
